# Logistic Regression

```python
import math
from copy import deepcopy

import numpy as np
import matplotlib.pyplot as plt

from mlutils import compute_cost_logistic, compute_gradient_logistic, gradient_descent_logistic, sigmoid
```

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])  #(m,n)
y_train = np.array([0, 0, 0, 1, 1, 1])   
```

```python
w, b, _ = gradient_descent_logistic(
    X_train,
    y_train,
    np.zeros_like(X_train[0]),
    0,
    0.1,
    10000,
    lambda_=0.7
)
```

```python
w, b
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

w = model.coef_
b = model.intercept_
n_iter = model.n_iter_

f"Model found optimal values w={model.coef_[0]}, b={b[0]}. Number of iterations: {n_iter[0]}"
```

```python
y_train_pred = model.predict(X_train)
```

```python
acc_train = model.score(X_train, y_train_pred)
acc_train
```

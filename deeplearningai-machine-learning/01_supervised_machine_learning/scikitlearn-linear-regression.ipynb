# Linear Regression in Scikit Learn

**LinearRegression()**: This model **uses the normal equation instead of Gradient Descent**.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from lab_utils_multi import load_house_data
plt.style.use('./deeplearning.mplstyle')
np.set_printoptions(precision=2)
```

```python
X_train, y_train = load_house_data()
X_features = ['size_sqft','n_bedrooms','n_floors','age']
X_train.shape
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train);
```

```python
w = linear_model.coef_
b = linear_model.intercept_
print(f"w = {w:}, b = {b:0.2f}")
```

```python
a_house = np.array([1200, 3, 1, 40])   # Vector
a_house_reshaped = a_house.reshape(-1, 4)
a_house.shape, a_house_reshaped.shape
```

```python
pred_wb = w @ a_house + b
pred = linear_model.predict(a_house_reshaped)
pred_wb == pred[0]
```

# Derivatives and Backpropagation


```python
from sympy import symbols, diff
```

```python
# Let's derive J = w**2

J, w = symbols('J, w')
J = w**2
J
```

```python
dJ_dw = diff(J, w)
dJ_dw
```

```python
# Calculate the derivative at point w=n
dJ_dw_2 = dJ_dw.subs([('w', 2)])
dJ_dw_3 = dJ_dw.subs([('w', 3)])
dJ_dw_4 = dJ_dw.subs([('w', 4)])

dJ_dw_2, dJ_dw_3, dJ_dw_4 # 4 (slope of J at w=2), 6 (slope of J at w=3), 8 (slope of J at w=4)
```

```python
# Arithmetically, the derivative is HOW MUCH the output of a function J(w) changes for small changes in the input variable w

w = 3
epsilon = 0.001
w_epsilon = w + epsilon

J_w = w**2
J_wepsilon = w_epsilon**2

# Amount of change = (after - before)/epsilon

dJ_dw = (J_wepsilon - J_w)/epsilon

dJ_dw
```

# Single variable Linear Regression

```python
from copy import deepcopy
import math

import numpy as np
import matplotlib.pyplot as plt

from lab_utils_uni import plt_gradients
```

```python
X_train = np.array([1.0, 2.0])  # in thousands of square feet
y_train = np.array([300.0, 500.0])
```

```python
from mlutils import compute_cost_v1
```

```python
from mlutils import compute_gradient_v1
```

```python
from mlutils import gradient_descend_v1
```

```python
initial_w = 0
initial_b = 0
alpha = 1.0e-2
n_iter = 10000

final_w, final_b, j_history, p_history = gradient_descend_v1(
    X_train,
    y_train,
    initial_w,
    initial_b,
    alpha,
    n_iter
)

print(f'Optimal (w, b): ({final_w: 8.4f}, {final_b: 8.4f})')
```

```python
# Cost Vs. Iterations plot
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4))

# plot 1 shows the first 100 iterations
ax1.plot(j_history[:100]);
ax1.set_title('Cost Vs. Iterations (start)')
ax1.set_ylabel('Cost')
ax1.set_xlabel('Iteration step')


# plot 2 shows from 1000th iter to the end
ax2_x_len = 1000 + np.arange(len(j_history[1000:])) # from 1000 to 1000 + the length of (len(j_hist) - 1000)
ax2.plot(ax2_x_len, j_history[1000:])
ax2.set_title('Cost Vs. Iterations (end)')
ax2.set_ylabel('Cost')
ax2.set_xlabel('Iteration step')
```

```python
predict = lambda x_val: (final_w * x_val + final_b)
```

```python
print(f'Predicting with final weigths w={final_w:0.3f}, b={final_b:0.3f}')
for i in [1.0, 1.2, 2.0]:
    print(f'Predicted price for a {i*1000} sqft house: {predict(i)}')
```
